```python
# 현재 주로 function을 사용

def add(a, b):
    a = tf.convert_to_tensor(a, dtype = tf.float32)
    b = tf.convert_to_tensor(b, dtype = tf.float32)
    return a+b

print(add(3, 4.5).numpy())
print(add([1,3], [2,4]).numpy())
```

# Linear Regression

```python
# data
x_train = [1,2,3]
y_train = [1,2,3]

# random_normal은 우리가 아직 W, b를 모르기 때문에 rank가 1인 값을 랜덤하게 넣어주는 것.
W = tf.Variable(tf.random.normal([1]), name = 'weight')
b = tf.Variable(tf.random.normal([1]), name = 'bias')
```

tensorflow.GradientTape()는 자동 미분을 수행하기 위해 사용되는 도구. 주로 머신러닝 모델의 훈련과정에서 손실함수에 대한 모델 변수(가중치, 바이어스 등)의 그래디언트를 계산하는데 사용.

```python
# 경사하강... Minimize

optimizer = tf.optimizers.SGD(learning_rate = 0.01)

# 최적화 function 정의
def train_step(x,y):
    with tf.GradientTape() as tape:
        hypothesis = x_train * W + b
        
        # cost/loss function
        cost = tf.reduce_mean(tf.square(hypothesis - y_train))

    # gradient 계산
    gradients = tape.gradient(cost, [W, b])

    # 변수 업데이트
    optimizer.apply_gradients(zip(gradients, [W, b]))
    return cost

for step in range(2001):
    train_step(x_train, y_train)
    if step % 20 == 0:
        print(f"Step: {step}, Cost: {cost.numpy()}, W: {W.numpy()}, b: {b.numpy()}")
```
